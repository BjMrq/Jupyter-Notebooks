```python
import os
import boto3
import pandas as pd
import numpy as np
from dotenv import load_dotenv

# Load environement variables
load_dotenv()

campus_name = "InterDec"

path = os.getcwd()
```

```python
# Get id
aws_id = os.environ.get('AWSAccessKeyId')
aws_secret = os.environ.get('AWSSecretKey')

# Get Bucket
key='last_week_leads.csv'
bucket_name='lcieducationleads'


# Initialitiong connection with S3
client = boto3.client('s3', aws_access_key_id=aws_id,
                              aws_secret_access_key=aws_secret)

object = client.get_object(Bucket=bucket_name, Key=key)

df = pd.read_csv(object['Body'])
df.to_csv(path + "\\" +'all_last_week_leads.csv',
                 header=True, index=False, index_label=False)
df
```

```python
df = pd.read_csv(path + "\\" +'all_last_week_leads.csv')
df
```

```python
if campus_name == "LaSalle" or campus_name == "InterDec":
    if campus_name == "LaSalle":
        campus_name = ["LaSalle College", "LaSalle College Laval"]

    elif campus_name == "InterDec":
        campus_name = ["Inter-Dec College Montréal",
                       "Inter-Dec College Laval"]

    # Isoalating paid media source
    leads = df.loc[df['Latest Campus of Interest'].isin(campus_name)]

# Isolate campus's leads
else:
    leads = df.loc[df['Latest Campus of Interest'] == campus_name]

# Isoalating paid media source
leads = leads.loc[leads['GA Source'].isin(['facebook', 'google'])]
leads
```

```python
# Get leads per source
pivot = pd.pivot_table(
        leads,
        values='E-mail 1',
        index=['Program Choice 1', 'Latest Campus of Interest'],
        columns=['GA Source'],
        aggfunc=lambda x: len(x.unique()), fill_value=0, margins=True).reset_index()

pivot = pivot.sort_values(by=['Latest Campus of Interest', 'Program Choice 1'], ascending=False)

# Save the result
campus_name = 'LaSalle'

pivot.to_csv(path + "\\" + campus_name + '_last_week_leads.csv', index=False)

pivot
```
